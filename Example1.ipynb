```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

```

```python
#data csv load name of csv file is Book
data=pd.read_csv('Book.csv')
```

```python
#see the data set use shift+enter
data 
```

```python
#visualize the data 
#x axis videos and y axis views
plt.scatter(data.videos,data.views,color='purple')
plt.xlabel('number of videos')
plt.ylabel('total views')
```

```python
data.views
```

```python
data.videos
```

```python
#convert data into array
x=np.array(data.videos.values)
x
```

```python
#convert data into array
y=np.array(data.views.values)
y
```

```python
# Train the model
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
```

```python
test=np.array([45]).reshape((-1, 1))
result=model.predict(test)
result
```

```python
plt.scatter(data.videos, data.views, color='green')
m, c = np.polyfit(x, y, 1)
plt.plot(x, m * x + c)
plt.xlabel('Videos')
plt.ylabel('Views')
plt.show()
```
